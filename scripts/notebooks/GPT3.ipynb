```python
import pandas as pd


df_train = pd.read_csv("/content/drive/MyDrive/CS546/course-project/SemEval/data/data_en_subtask3/train.csv")
df_test = pd.read_csv("/content/drive/MyDrive/CS546/course-project/SemEval/data/data_en_subtask3/test.csv")
```

```python
df_train
```

```python
df_train[df_train['labels'].str.contains('Doubt')]
```

```python
st3_labels = set()

for labels in df_train["labels"].values:
    for label in labels.split(","):
        st3_labels.add(label)

st3_labels = list(st3_labels)

st3_labels
```

```python
# distribution of 
labels_freq = []
for i, label in enumerate(st3_labels):
    df_i = df_train[df_train['labels'].str.contains(label)]

    labels_freq.append((label, len(df_i)))

labels_freq.sort(key=lambda x: -x[1])

labels_freq
```

```python
for label, freq in labels_freq:
    if freq > 100:
        print(label)
```
