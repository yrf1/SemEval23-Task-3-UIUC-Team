```python
df = pd.read_csv("/content/drive/MyDrive/CS546/course-project/SemEval/data/data_en_subtask3/train.csv")

df
```

```python
st3_labels = set()

for labels in df["labels"].values:
    for label in labels.split(","):
        st3_labels.add(label)

st3_labels = list(st3_labels)

# distribution of the labels and sort by frequency
labels_freq = []
for i, label in enumerate(st3_labels):
    df_i = df[df['labels'].str.contains(label)]

    labels_freq.append((label, len(df_i)))

labels_freq.sort(key=lambda x: -x[1])

labels_freq
```
